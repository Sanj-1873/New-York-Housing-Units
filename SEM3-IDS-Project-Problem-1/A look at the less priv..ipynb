```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", color_codes=True)
```

```python
# Reading the CSV File 
df = pd.read_csv('dataset.csv')
print(df.columns)
#print(df.extremely_low_income_units)
#df
Total_r = df['counted_rental_units'].sum()
print (Total_r)
Total_o = df['counted_homeownership_units'].sum()
print (Total_o)
Total_t = df['total_units'].sum()
print (Total_t)
df2 = pd.DataFrame({'total': [Total_r, Total_o]},index=['counted_rental_units', 'counted_homeownership_units'])
plot = df2.plot.pie(y='total', figsize=(5, 5))
```

```python
#extended_affordability_only
df_eao = df.loc[df.extended_affordability_only == 'Yes']
print('Num of extended affordability units',len(df_eao))
#Comparing size and location 
eao_br1 = sns.catplot(x="borough", y="br1_units", kind="swarm", data=df_eao);
eao_br2 = sns.catplot(x="borough", y="br2_units", kind="swarm", data=df_eao);
eao_br3 = sns.catplot(x="borough", y="br3_units", kind="swarm", data=df_eao);
eao_br4 = sns.catplot(x="borough", y="br4_units", kind="swarm", data=df_eao);
eao_br5 = sns.catplot(x="borough", y="br5_units", kind="swarm", data=df_eao);
eao_br6 = sns.catplot(x="borough", y="br6_units", kind="swarm", data=df_eao);
eao_bruk = sns.catplot(x="borough", y="unknown-br_units", kind="swarm", data=df_eao);

#Comparing renting and ownership with the location
eao_br_rent = sns.catplot(x="borough", y="counted_rental_units", kind="swarm", data=df_eao);
eao_br_own = sns.catplot(x="borough", y="counted_homeownership_units", kind="swarm", data=df_eao);
```
