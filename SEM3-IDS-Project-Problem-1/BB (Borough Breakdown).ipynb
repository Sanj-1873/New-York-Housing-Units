```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Reading the CSV File 
df = pd.read_csv('dataset.csv')
print(df.columns)
print(df.postcode)
```

```python
## All things Bronx 
df_Bronx = pd.DataFrame(df.loc[df['borough']=="Bronx"])
#print(df_Bronx.columns)
print("Total number of projects started in the Bronx",len((df_Bronx).project_id))
b1 = sum(df_Bronx.br1_units)
b2 = sum(df_Bronx.br2_units)
b3 = sum(df_Bronx.br3_units)
b4 = sum(df_Bronx.br4_units)
b5 = sum(df_Bronx.br5_units)
b6 = sum(df_Bronx.br6_units)
everything = b1 + b2 + b3+ b4 +b5 +b6 
x = ['b1','b2','b3','b4','b5','b6']
y = [b1,b2,b3,b4,b5,b6]

print("Total number of 1_br_units",b1)
print("Total number of 2_br_units",b2 )
print("Total number of 3_br_units", b3)
print("Total number of 4_br_units", b4)
print("Total number of 5_br_units", b5)
print("Total number of 6_br_units", b6)

df_bx = pd.DataFrame({'size':['b1','b2','b3','b4','b5','b6'], 'num':[b1,b2,b3,b4,b5,b6]})
ax = df_bx.plot.bar(x='size', y='num', rot=0)

Total_e_l = df_Bronx['extremely_low_income_units'].sum()
print (Total_e_l)
Total_v_l = df_Bronx['very_low_income_units'].sum()
print (Total_v_l)
Total_l = df_Bronx['low_income_units'].sum()
print (Total_l)
Total_mo = df_Bronx['moderate_income_units'].sum()
print (Total_mo)
Total_mid = df_Bronx['middle_income_units'].sum()
print (Total_mid)
df_Bronx_2 = pd.DataFrame({'total': [Total_e_l, Total_v_l,Total_l,Total_mo,Total_mid]},index=['extremely_low_income_units', 'very_low_income_units','low_income_units','moderate_income_units','middle_income_units'])
plot = df_Bronx_2.plot.pie(y='total', figsize=(5, 5))

cd = df_Bronx.council_district.unique().tolist()
print(cd)
for dit_num in cd:
    test = df.loc[df['council_district']==dit_num]
    #print(test)
    Total_e_l = test['extremely_low_income_units'].sum()
    
    Total_v_l = test['very_low_income_units'].sum()
    
    Total_l = test['low_income_units'].sum()
   
    Total_mo = test['moderate_income_units'].sum()
 
    Total_mid = test['middle_income_units'].sum()
    
    df2 = pd.DataFrame({'total': [Total_e_l, Total_v_l,Total_l,Total_mo,Total_mid]},index=['extremely_low_income_units', 'very_low_income_units','low_income_units','moderate_income_units','middle_income_units'])
    plot = df2.plot.pie(y='total', figsize=(5, 5))
    print('The distict zone we are looking at here is', dit_num)

```
